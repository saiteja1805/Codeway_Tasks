```python
dataset = pd.read_csv("/content/drive/MyDrive/codeway_datasets/Churn_Modelling.csv")
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
dataset["Surname"] = LabelEncoder().fit_transform(dataset["Surname"])
dataset["Geography"] = LabelEncoder().fit_transform(dataset["Geography"])
dataset["Gender"] = LabelEncoder().fit_transform(dataset["Gender"])
```

```python
dataset.info()
```

```python
exited = dataset[dataset.Exited==1]
not_exited = dataset[dataset.Exited==0]
new_dataset = pd.concat([not_exited,exited],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
from sklearn.model_selection import train_test_split
x = new_dataset.iloc[:,:-1]
y = new_dataset["Exited"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
x_pred = model.predict(x_train)
```

```python
# training_accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(x_pred,y_train)
train_accuracy
```

```python
y_pred = model.predict(x_test)
```

```python
# testing_accuracy
testing_accuracy = accuracy_score(y_pred,y_test)
testing_accuracy
```
